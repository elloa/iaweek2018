```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("autompg.csv",sep = ";")
df.head()
```

```python
df.loc[df['name'] == "chevrolet camaro"]
```

```python
np.mean(df["mpg"])
```

```python
df.dropna(inplace=True)
df.drop(['name'],axis = 1,inplace=True)
df.head()
```

```python
kml = [round(x*0.425,2) for x in df["mpg"]]
np.mean(kml)
```

```python
df['kml'] = kml
df.drop(['mpg'],axis = 1)
```

```python
Y = df['kml']
X = df.drop(["kml"],axis = 1)
X.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30)
```

```python
regr = LinearRegression()

# Train the model using the training sets
regr.fit(X_train, Y_train)
```

```python
teste1 = X_test[2:3]
resultado = regr.predict(teste1)
print(resultado,Y_test.iloc[2])
#erro = abs(resultado - Y_test[0])
#erro
```

```python
Y_predito = regr.predict(X_test)
rmse = mean_squared_error(Y_predito,Y_test)
rmse
```

```python
r2 = r2_score(Y_predito,Y_test)
r2
```

```python
Y_test = pd.Series.tolist(Y_test)
Y_predito = Y_predito.tolist()
```

```python
residuos = []
for i in range(len(Y_test)):
    residuos.append((Y_test[i] - Y_predito[i])**2)
residuos
```

```python
x = [0,int(max(Y_test))]
y = [0,0]
plt.plot(x,y,linewidth=3)
plt.plot(Y_test,residuos,'ro')
plt.ylabel('Residuos')
plt.xlabel('kml')
plt.show()
```

```python
kviz = KNeighborsRegressor(n_neighbors=21)
kviz.fit(X_train,Y_train)
```

```python
results = kviz.predict(X_test)
```

```python
rmse2 = mean_squared_error(results,Y_test)
rmse2
```

```python
r22 = r2_score(results,Y_test)
r22
```
